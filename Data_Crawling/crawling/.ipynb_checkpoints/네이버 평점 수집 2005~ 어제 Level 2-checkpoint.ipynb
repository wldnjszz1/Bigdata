```python
'''
평점 별 랭킹 첫 집계일
https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20050207

현재 시점까지 집계일 (오늘 -1)
https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190210
'''
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url_base = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date='
page = urlopen( url_base + '20190210')
soup = BeautifulSoup(page,'html.parser')
soup
```

```python
# 날짜 20050207 ~ 20190210 연속된 문자열
import pandas as pd
# 시간의 연속 값
# dates = pd.date_range('2005-2-7', periods=5117, freq='D')
dates = pd.date_range('2019-2-9', periods=2, freq='D')
print(dates[:1], dates[-1:], len(dates))
```

### 데이터 수집 필드
- 평점일(날짜, regdate timestamp)
- 영화제목 title varchar
- 평점 score float
- 출처 site(값 naver) varchar

```python
for date in dates:
    # print(str(date)[:10].replace('-','')) # timestamp 문자열화해서 처리한 방식
    # print(date.strftime('%Y%m%d')) # timestamp 자체에서 날짜 
    page = urlopen( url_base + date.strftime('%Y%m%d') )
    soup = BeautifulSoup(page,'html.parser')
    # 파싱 => 적재
    # 평점일 : date
    # 영화제목 : .tit5
    titles = soup.select('.tit5')
    # 평점 : .point
    points = soup.find_all('td','point') # td들 중에 class가 point인 모든 요소
    # print(titles[0].a.string, points[0].string, '데이터 개수는 일치 하는가 ? ', len[titles]==len(points))
    
    
```

```python
# 진행율 표시 (오래 걸릴 경우 진행율을 시각적으로 확인)
from tqdm import tqdm_notebook
tmp = list()
dates = pd.date_range('2005-2-7', periods=5117, freq='D')
for date in tqdm_notebook(dates):
    page   = urlopen( url_base + date.strftime('%Y%m%d') )
    soup   = BeautifulSoup(page,'html.parser')
    titles = soup.select('.tit5')
    points = soup.find_all('td','point') 
    dump   = [ {
                'title'  :titles[i].a.string,
                'score'  :float(points[i].string),
                'site'   :'naver',
                'regdate':date
                } for i in range(len(titles)) ]
    tmp.extend(dump)
    
```

```python
# 디비에 밀어 넣기
import pymysql 
from sqlalchemy import create_engine
import pandas as pd
import pandas.io.sql as pSql

df_dict = pd.DataFrame.from_dict( tmp )

db_url = 'mysql+pymysql://root:1234@127.0.0.1/python_db'
engine = create_engine(db_url, encoding = 'utf8')
conn = engine.connect()
df_dict.to_sql( name = 'movie_rank', con=conn, if_exists='append', index = False )
```
