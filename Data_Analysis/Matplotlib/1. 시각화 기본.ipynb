```python
# 모듈 import
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm # 한글이 안깨지기 위해
# 버전이 낮은 jupyter를 사용할 경우 => %matplotlib inline

# 현재 컴퓨터에서 사용 가능한 폰트 목록을 가져온다
# font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')

# 폰트 설정
font_name = 'c:/Windows/Fonts/HMFMPYUN.TTF'
font_pro = fm.FontProperties(fname=font_name)
font_family = font_pro.get_name()

plt.rcParams['font.family'] = font_family
```

```python
# 라인
# 가장 간단한 그래프로 선을 그리는 라인 그래프
# 시간, 순서에 따라 어떻게 변화하는지 보여주기 위해 사용된다
plt.title('Line Plot')

# 첫번째 : x축 좌표 값, 두번째 : y축 좌표 값
plt.plot([100, 200, 300, 400],[1, 4, 9, 16])
plt.show()
```

```python
# 스타일
# 색상 + 마커 + 선종류 (r: red, s: 마커의 종류, --: 선)
# 색상 : https://matplotlib.org/examples/color/named_colors.html
# 마커 : https://matplotlib.org/gallery/lines_bars_and_markers/marker_reference.html#sphx-glr-gallery-lines-bars-and-markers-marker-reference-py\n
# 선 : https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html#sphx-glr-gallery-lines-bars-and-markers-line-styles-reference-py\n
plt.title('스타일')
plt.plot([100, 200, 300, 400], [1, 4, 9, 16], 'bv-.') 
plt.show()
```

```python
plt.title('개별 설정')
# c : 색상, lw : 선 굵기, ls : 선 스타일
# marker : 마커종류, ms : 마커크기
# mec : 마커 선 색깔, mew : 마커 선 굵기
# mfc : 마커 내부 색깔
plt.plot([100, 200, 300, 400], [1, 4, 9, 16], c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

```python
# x, y축 범위 설정
plt.title('x축, y축 범위 설정')
plt.plot([100, 200, 300, 400], [1, 4, 9, 16])
plt.xlim(0, 1000)
plt.ylim(-10, 50)
plt.grid(True) # 그리드 표시
plt.show()
```

```python
# 점 그래프
plt.title('점 그래프')
plt.scatter([100, 200, 300, 400], [1, 4, 9, 16])
plt.show()
```

```python
# 바 차트
x = ['A', 'B', 'C', 'D']
y = [1, 4, 9, 16]
y2 = [5, 10, 4, 8]
plt.title('바 차트')
plt.bar(x, height=y, width=0.1)
plt.bar(x, height=y2, width=0.1, bottom=y)
plt.show()
```
