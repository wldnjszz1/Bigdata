# 데이터 생성 함수

주어진 조건을 이용해 값을 생성하여 행렬을 만든다

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# linspace : start부터 stop까지의 범위에서 N개의 값을 생성하여 행렬을 만든다

shape1 = np.linspace(0, 1, 5)
print(f'shape1 : {shape1}')

shape2 = np.linspace(0, 10, 5, dtype=int)
print(f'shape2 : {shape2}')

plt.plot(shape1,'o')
plt.show()
plt.plot(shape2,'v')
plt.show()
```

```python
# arange : start부터 stop미만까지 범위에서 step씩 증가하는 행렬을 만든다
shape1 = np.arange(0, 1, 0.1)
print(f'shape1 : {shape1}')

shape2 = np.arange(0, 10, 1.5, dtype=int) # 증가량과 type값이 다르면 엉뚱한 값이 나옴
print(f'shape2 : {shape2}')
```

```python
# logspace : start부터 stop까지의 N개의 개수만큼 균등 간격으로 데이터를 생성하고 배열을 만든다
# 값은 log scale값
shape1 = np.logspace(0.1, 1, 20)
print(f'shape1 : {shape1}')

plt.plot(shape1)
plt.show()
```

```python
# 랜덤값 구하기
shape1 = np.random.random()
print(f'shape1 : {shape1}')

shape2 = np.random.random(size=5)
print(f'shape2 : {shape2}')

shape3 = np.random.random(size=(2,2))
print(f'shape3 : {shape3}')

plt.plot(shape1)
plt.plot(shape2)
plt.plot(shape3)
plt.show()
```

```python
# 추출되는 랜덤값에 정규분포를 적용한다
# mean : 평균
# std : 표준편차
mean = 0
std = 1
shape1 = np.random.normal(mean, std, 5)
print(f'shape1 : {shape1}')

shape2 = np.random.normal(mean, std, (2,2))
print(f'shape2 : {shape2}')

plt.plot(shape1)
plt.plot(shape2)
plt.show()
```

```python
# 정수 랜덤값 행렬 생성
# 범위 : min ~ max -1
min = 3
max = 14

shape1 = np.random.randint(min, max, 5)
print(f'shape1 : {shape1}')

shape2 = np.random.randint(min, max, (3, 3))
print(f'shape2 : {shape2}')
```
